```python
# Slicing (Rebanado)
frase = "Python Programing"
# Sintaxis: String[Inicio:Fin:Paso]
```

```python
print(len(frase))
print(frase[0]) # imprime el primer caracter
print(frase[16]) # imprime el ultimo caracter
print(frase[-1]) # imprime el ultimo caracter
print(frase) # imprime la frase completa
print(frase[0:16]) # imprime desde el indice 0 hasta el 15
print(frase[0:17]) # imprime desde el indice 0 hasta el 16
print(frase[:17]) # imprime desde el inicio hasta el indice 16
print(frase[7:]) # imprime desde el indice 7 hasta el final
```
