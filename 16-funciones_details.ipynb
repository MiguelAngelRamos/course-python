```python
def multiplicar(*args):
    # llega en forma tupla (20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10)
    resultado = 1
    for numero in args:
        resultado = resultado * numero # y lo podemos abreviar resultado *= numero
    return resultado

resultado_multiplicacion = multiplicar(20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10)
print(resultado_multiplicacion)
```

```python
def crear_perfil(**kwargs):
    # kwargs llega en forma de diccionario {'nombre': 'Sofia', 'edad': 30, 'profesion': 'Desarrolladora', 'ciudad': 'Madrid'}
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")
    

crear_perfil(nombre="Sofia", edad=30, profesion="Desarrolladora", ciudad="Madrid")
print("\n") # para separar los resultados
crear_perfil(nombre="Catalina", profesion = "Diseñadora", web="www.catalina.com", twitter="@catalina", instagram="@catalina_insta", linkedin="@catalina_linkedin", github="@catalina_github")
```

```python
# Ejemplo valores por defecto con **kwargs

def registrar_usuario(**kwargs):
    # kwargs llega en forma de diccionario {'nombre': 'Sofia', 'rol': 'Admin'}
    # Puedo usar el metodo get para obtener los valores de las claves, y si no existen, asignar un valor por defecto
    nombre = kwargs.get('nombre', 'Usuario desconocido')
    rol = kwargs.get('rol', 'Rol no especificado')
    print(f"Registrando usuario: {nombre} con rol: {rol}")


registrar_usuario(nombre="Sofia", rol="Admin")
print("\n")
registrar_usuario(nombre="Richard")
print("\n")
registrar_usuario(rol="User")
print("\n")
registrar_usuario()
print("\n")
registrar_usuario(nombre="Carlos")
```

```python

def conectar_base_datos(**kwargs):
    # kwargs llega en forma de diccionario {'host': 'localhost', 'puerto': 5432, 'usuario': 'admin', 'contraseña': 'admin123', 'base_datos': "tech_db"}
    print("Conectando a la base de datos con la siguiente configuración...")
    print(f"Conectando a {kwargs['host']}:{kwargs['puerto']} -> Base de datos: {kwargs['base_datos']}")

# Esto es un diccionario con la configuración de la base de datos, que podría ser utilizado para conectar a una base de datos
configuracion_base_datos = {
    'host': 'localhost',
    'puerto': 5432,
    'usuario': 'admin',
    'contraseña': 'admin123',
    'base_datos': "tech_db"
}
# Podemos usar el operador de desempaquetado ** para pasar el diccionario como argumentos a la función
conectar_base_datos(**configuracion_base_datos)
# conectar_base_datos(host='localhost', puerto=5432, usuario='admin', contraseña='admin123')
```
