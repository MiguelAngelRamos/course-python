```python
def sumar_todos(*args):
  return sum(args)

resultado_suma = sumar_todos(60, 100, 50, 30, 25)
print(resultado_suma)
```

```python
def presentar_equipo(*miembros):
    for miembro in miembros:
        print(f"Miembro del equipo: {miembro}")

presentar_equipo("Sofia", "Carlos", "Ana", "Luis", "Catalina")
```

```python
def multiplicar(*args):
    # llega en forma tupla (20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10)
    resultado = 1
    for numero in args:
        resultado = resultado * numero # y lo podemos abreviar resultado *= numero
    return resultado

resultado_multiplicacion = multiplicar(20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10, 30, 20, 10, 5, 10, 5, 10)
print(resultado_multiplicacion)
```

```python
def crear_perfil(**kwargs):
    # kwargs llega en forma de diccionario {'nombre': 'Sofia', 'edad': 30, 'profesion': 'Desarrolladora', 'ciudad': 'Madrid'}
    for clave, valor in kwargs.items():
        print(f"{clave}: {valor}")
    

crear_perfil(nombre="Sofia", edad=30, profesion="Desarrolladora", ciudad="Madrid")
print("\n") # para separar los resultados
crear_perfil(nombre="Catalina", profesion = "Diseñadora", web="www.catalina.com", twitter="@catalina", instagram="@catalina_insta", linkedin="@catalina_linkedin", github="@catalina_github")
```
