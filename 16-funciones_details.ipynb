```python
def sumar_todos(*args):
  return sum(args)

resultado_suma = sumar_todos(60, 100, 50, 30, 25)
print(resultado_suma)
```

```python
def presentar_equipo(*miembros):
    for miembro in miembros:
        print(f"Miembro del equipo: {miembro}")

presentar_equipo("Sofia", "Carlos", "Ana", "Luis", "Catalina")
```
