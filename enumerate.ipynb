```python
# Servicios de programaci√≥n

servicios = {
    "web": "React",
    "Movil": "Flutter",
    "Backend": "Node.js"
}

tecnologias = ["React", "Flutter", "Node.js"]
```

```python
for indice, tecnologia in enumerate(tecnologias):
    print(f"{indice}: {tecnologia}")
```
